```python
# Sol: 1

def check_squence(keyStroke):
    sequence = {}
    skip_to = 0
    for index in range(0, len(keyStroke)):
        if index == skip_to:
            count = get_count(index, keyStroke)
            skip_to = index + count
            sequence['{}-{}'.format(keyStroke[index], index)] = count - 1
    return sequence


def get_count(index, keyStroke):
    char = keyStroke[index]
    count = 1
    for nextIndex in range(index+1, len(keyStroke)):
        if char is keyStroke[nextIndex]:
            count = count + 1
        else:
            return count
    return count


def numbers_to_chars():

    keys = {
        '2': ['a', 'b', 'c'],
        '3': ['d', 'e', 'f'],
        '4': ['g', 'h', 'i'],
        '5': ['j', 'k', 'l'],
        '6': ['m', 'n', 'o'],
        '7': ['p', 'q', 'r', 's'],
        '8': ['t', 'u', 'v'],
        '9': ['w', 'x', 'y', 'z'],
    }

    while True:
        keyStroke = input('Enter the keystroke:')
        if len(keyStroke) > 0:
            if keyStroke.isdecimal():
                sequence = check_squence(keyStroke)
                message = ''
                for key, index in sequence.items():
                    char = key[0: key.find('-')]
                    message = message + keys[char][index]
                print(message)
            else:
                print('please enter a valid keystroke')
        else:
            print('please enter atleast one keystroke')


numbers_to_chars()
```

```python
# Sol: 4

open_list = ["[","{","("]
close_list = ["]","}",")"]

stack=[]
def check_parentheses(input_str):
    for i in input_str:
        if i in open_list:
            stack.append(i)
        elif i in close_list:
            pos = close_list.index(i)
            if ((len(stack) > 0) and
                (open_list[pos] == stack[len(stack)-1])):
                stack.pop()
            else:
                return False
    if len(stack) == 0:
        return True
    else:
        return False
   
input_str = input("Enter the Paranthesis")
check_parentheses(input_str)
```

```python
# Sol: 10

def adj_sum_even(n, Numbers):
   count = 0
   list1=[]

   for i in range(n):
       if (Numbers[i] % 2):
           count += 1
           list1.append(Numbers[i])
   
   return (min(count, n - count),list1)
 
Numbers=[1,3,5,4,2]
n = len(Numbers)
print(adj_sum_even(n, Numbers))
```
